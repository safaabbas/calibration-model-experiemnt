```python
import pandas as pd 
```

```python
path_to_kaggle = 'classification-dataset-v1.csv' # EDIT WITH YOUR FILE PATH 

df = pd.read_csv(path_to_kaggle,error_bad_lines=False)
df.head()
```

```python
downsample_by = 0.1 # Takes 10% of each class you can edit this to change the size of your dataset subset 
subset_df = df.groupby('Category', group_keys=False).apply(lambda x: x.sample(frac=downsample_by, random_state=42))
```

```python
subset_df['Category'].value_counts().plot(kind='bar')
```

```python
path_to_save = 'subset_kaggle.csv' # EDIT WITH YOUR FILE PATH 
subset_df.to_csv(path_to_save)
```
